```python
import pandas as pd
import random

# Function to add a random emoji to a text
def add_random_emoji(text):
    emojis = ["üòÄ", "üòÇ", "üòé", "üëç", "‚ù§Ô∏è", "üéâ","üòò"," ü´¢","üò¢","üò°","üëä","ü§ï"]
    return text + random.choice(emojis)

# Read the CSV dataset
csv_file_path = "concatenated.csv"
df = pd.read_csv(csv_file_path)

# Apply the add_random_emoji function to a column (e.g., "text_column") in your DataFrame
df["Comment"] = df["Comment"].apply(add_random_emoji)

# # Save the modified DataFrame back to a CSV file

# df.to_csv(output_csv_file_path, index=False)

# print(f"Modified dataset saved to {output_csv_file_path}")
```

```python
import random
import string

def add_punctuation_noise(text):
    punctuation_marks = string.punctuation  # Includes all punctuation marks
    noisy_text = list(text)
    for _ in range(3):  # You can control how many punctuations to add
        position = random.randint(0, len(noisy_text) - 1)
        noisy_text[position] = random.choice(punctuation_marks)
    return "".join(noisy_text)

text = "This is a sentence."
noisy_text = add_punctuation_noise(text)
print(noisy_text)
```
